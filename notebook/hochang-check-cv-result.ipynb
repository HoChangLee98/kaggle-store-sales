```python
import sys
sys.path.append("../src")

from validation import *

import matplotlib.pyplot as plt
```

```python
def check_cv_result(version:str):
    validation = Validator(folder_path="../artifacts")
    cv_result = validation.load_cv_result(version=version)
    plt.figure(figsize=(12, 5))
    plt.scatter(x=np.arange(len(cv_result["y_valid"])), y=cv_result["y_valid"], label="true", s=0.5)
    plt.scatter(x=np.arange(len(cv_result["y_valid"])), y=cv_result["y_valid_pred"], label="pred", s=0.5)
    plt.title(f"Cross Validation {version} Result")
    plt.legend()
    plt.show()
```

```python
check_cv_result(version="baseline_0")
```
