```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import matplotlib.pyplot as plt

from darts import TimeSeries
from darts.models import DLinearModel, RNNModel, ExponentialSmoothing, BlockRNNModel

from tqdm.auto import tqdm
from sklearn.metrics import mean_squared_log_error
```

```python
train = pd.read_csv("../data/train.csv")
train["family_store"] = train["family"] + "_" + train["store_nbr"].astype(str)

test = pd.read_csv("../data/test.csv", index_col='date', parse_dates=True)
test["family_store"] = test["family"] + "_" + test["store_nbr"].astype(str)

train = train.rename(columns={"date":"ds", "sales":"y"}).drop(columns=["id", "family", "store_nbr"])
test = test.rename(columns={"date":"ds", "sales":"y"}).drop(columns=["id", "family", "store_nbr"])
```

```python
period = (pd.Timestamp("2013-01-01"), pd.Timestamp("2016-08-15"), pd.Timestamp("2016-08-31"))
train["ds"] = pd.to_datetime(train["ds"])
train_set = train[(train["ds"] >= period[0]) & (train["ds"] <= period[1])]
valid_set = train[(train["ds"] > period[1]) & (train["ds"] <= period[2])]
```

```python
random_seed = 0

def trainer(train_set):
    pred = {}
    for fs in tqdm(train_set["family_store"].unique()):
        single = train_set[train_set["family_store"] == fs]
        single_ts = TimeSeries.from_dataframe(single[["y"]])
        model = DLinearModel(
            input_chunk_length=128, 
            output_chunk_length=28, 
            batch_size=128, 
            n_epochs=20
            )
        model.fit(single_ts)
        
        future = pd.DataFrame()
        future['ds'] = pd.date_range(start='2016-08-16',periods=16,freq='D')
        forecast = model.predict(len(future['ds']))
        y_pred = forecast.values()
        pred["fs"] = y_pred
    return pred
```

```python
forecast = trainer(train_set)
```

```python
mean_squared_log_error(valid_set[valid_set["family_store"] == "AUTOMOTIVE_1"].y, forecast.values())
```
